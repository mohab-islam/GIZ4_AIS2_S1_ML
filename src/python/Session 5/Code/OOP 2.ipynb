Abstract Class

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    
    @abstractmethod
    def move(self):
        pass

class bird(Animal):
    pass

class cat(Animal):

    def move(self):
        print("Cat")
```

Multi-inheritence

```python
class A():
    def do_this(self):
        print("Doing this in A")
class B(A):
    pass
class C():
    def do_this(self):
        print("Doing this in C")
class D(C,B):
    pass

d = D()
d.do_this()
```

```python
class A():
    def do_this(self):
        print("Doing this in A")
class B(A):
    pass
class C():
    def do_this(self):
        print("Doing this in C")

class D(B,C):
    pass

dd = D()
dd.do_this()
```

```python
listofwords = ["Ali", "Mohab", "Omar", "Youssef", "Ahmed", "Ali", "Mohab", "Omar", "Youssef", "Ahmed", "Mohab"]
def unique_words(words):
    counts = {}
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts
```

```python
unique_words(listofwords)
```

```python
import random

def get_response(user_input):
    responses_list = {
        "hello": ["hello", "hi", "hey"],
        "how are you?": ["I'm fine, thank you.", "Doing well, and you?", "Great! How about you?"],
        "goodbye": ["See you later!", "Goodbye!", "Take care!"],
        "default": ["I'm sorry, I don't understand."]
    }
    
    for key in responses_list:
        if key in user_input:
            return random.choice(responses_list[key])
    
    return random.choice(responses_list["default"])

def chatbot():
    print("chatbot: Hi, how can I help you today?")
    
    while True:
        user_input = input("user: ").lower()
        response = get_response(user_input)
        print("chatbot: ", response)

        if user_input == "goodbye":
            break

chatbot()
```

```python
def fact(num):
    if num == 0:
        return 0
    if num == 1:
        return 1
    return num*fact(num -1 )
```

```python
fact(5)
```

```python
def isprim(x):
    for i in range(2, x):
        if x % i == 0:
            return False
        return True
isprim(6)
```

```python
def dividors(num1, num2):
    """
    param1 = enter the first number of the common divisors
    type:param1 = int

    param2 = enter the 2nd number of the common divisors
    type:param2 = int

    return: the common dividors
    type_return: int
    """
    mn = min(num1, num2)
    if num1 == num2:
        print("please enter different nums")
        return
    CD = []
    for i in range(1,mn+1):
        if num1 % i == 0 and num2 % i == 0:
            CD.append(i)
    return CD
```

```python
dividors(10,10)
```

```python
def even_fun(num):
    eveen = []
    for i in range(num):
        if i % 2 == 0:
            eveen.append(i)
    return eveen
```

```python
even_fun(5)
```

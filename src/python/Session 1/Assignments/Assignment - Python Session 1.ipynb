Task 1

```python
email = "m.islam2186@nu.edu.eg"
#input validation
num_of_at = email.count("@")
num_of_dot = email[email.index("@"):].count(".")
is_valid = (num_of_at == 1) and (num_of_dot >= 1)
print("Email is Valid: " , is_valid)
print("Username: ", email[:email.index("@")])
print("Domain: ", email[email.index("@")+1:])
#I've inculded the last dot because in case of Edu email, the last part matters
domain = email[email[email.index("@"):].index(".")+1:]
Commercial = "com" in domain
Edu = "edu" in domain
Other_domain =  "com" not in domain and "edu" not in domain
print("Domain:-", "Commercial"*Commercial, "Edu"*Edu , "Other Domain"*Other_domain)
```

Task 2

```python
Message = "###!!@emocleW PGTQ!!!6789"
#Decoding the Message
#Step-1, extract the core part
CORE = Message[Message.index("@")+1:Message.index("Q")+1]
first_word = CORE[:CORE.index("W")+1]
print(first_word)
#Step-2, Reverse first word
first_word_reversed = first_word[::-1]
print(first_word_reversed)
seond_word = CORE[CORE.index(" ")+1:]
print(seond_word)
#Step-3, print final msg
print(first_word_reversed + " " + seond_word)

```

Task 3

```python
Message2 = "&&&**$gnirts PLIO!!1234"
#Decoding the Message
#Step-1, extract the core part
CoreMSG= Message2[Message2.index("$")+1:Message2.index("!")]
print(CoreMSG)
#Step-2, Reverse first word
firstWordRev = CoreMSG[CoreMSG.index("s")::-1]
print(firstWordRev)
#Step-3, Replace shifted vowels in the second word
secWord = CoreMSG[CoreMSG.index("P"):]
secWordReplaced1 = secWord.replace("I","E")
FinalSecWord = secWordReplaced1.replace("O","U")
print(FinalSecWord)
#Step-4, print final msg
print(firstWordRev, FinalSecWord)
```

```python
Message3 = "##$$$@!yalpstcejorp EPUVT****9887"
#Decoding the Message
#Step-1, extract the core part
CoreMSG2= Message3[Message3.index("!")+1:Message3.index("*")]
print(CoreMSG2)
#Step-2, Reverse first word
firstWordRev = CoreMSG2[CoreMSG2.index(" ")::-1]
print(firstWordRev)
#Step-3, Replace shifted vowels in the second word
secWord2 = CoreMSG2[CoreMSG2.index("E"):]
secWordReplaced2 = secWord2.replace("E","A")
FinalSecWord2 = secWordReplaced2.replace("U","O")
print(FinalSecWord2)
#Step-4, print final msg
print(firstWordRev, FinalSecWord2)
```
